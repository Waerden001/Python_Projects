## Solution

We clicked around the downloadable pdf links, ** a pattern appears**, a typical link looks like ```http://d2aohiyo3d3idm.cloudfront.net/publications/virtuallibrary/0892363657.pdf```, all such links have most of their characters the same, actually ```http://d2aohiyo3d3idm.cloudfront.net/publications/virtuallibrary/089236``` always appears, the last four digits vary, when we apply ```request```, the returned response has a status code,  most of these links don't exist at all, so they return ```404``` or similar errors, when it's a downloadable pdf, the code is ```200```. Thus, we
- enumerate all numbers smaller thatn $10000$ in a 4-digit form, and try all link = base_link+ varing 4 digits+.pdf, that is ```res = requests.get(link)```
- check ```res.status_code```, if it's ```200```, we add the link to our content list
- download from the content list with ``` urllib.request```, **this fails in local jupyter notebook, but works smoothly with google colab, somehow**
